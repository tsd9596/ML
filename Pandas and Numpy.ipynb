### iloc and loc

#### loc

- It is label based data selecting method which means we have to pass the name of col which we want to select.
- This method included that last element of the range passed in it.
- Range excluded.

```python
df.loc[:,'Gender':'Quantity']
```

```python
df.loc[:,'Customer_id':'Price']
```

```python
df.loc[:,'Age':'Quantity']
```

#### iloc()

- It is index based data selecting method which means we have to pass integer index of spacified col.
- Range included.

```python
df.iloc[:,1:5]
```

```python
df.iloc[:,0:6]
```

```python
df.iloc[:,1:5]
```

### Concatining

- Concat
- Merge

```python
dfc1 = pd.DataFrame({'EMP_ID':[101,102,103,104,105],
                     'EMP_NAME' :['Isha','Priya','Neha','Rahul','Abhishek'],
                     'EMP_DEPT' :['E-101','E-104','E-101','E-102','E-101'],
                     'CONTACT_NO':[1234567890,1234567890,1234567890,1234567890,1234567890],
                     'EMAIL_ID' :['isha@gmail.com','priya@yahoo.com','neha@gmail.com','rahul@yahoo.com','abhishek@gmail.com'],
                     'EMP_HEAD_ID' :[105,103,101,105,102]})
```

```python
dfc1
```

```python
dfc2 = pd.DataFrame({'EMP_DEPT' :['E-101','E-102','E-103','E-104','E-105'],
                     'DEPT_NAME':['HR','Development','Hous Keeping','Sales','Purchage'],
                     'DEPT_OFF':['Monday','Tuesday','Saturday','Sunday','Tuesday',],
                     'EMP_ID':[105,101,103,102,104]
                    })
```

```python
dfc2
```

```python
pd.concat([dfc1,dfc2],ignore_index=True)
```

```python
pd.concat([dfc1,dfc2],axis=1)
```

```python
pd.concat([dfc1,dfc2],keys=['Employee','Department'])
```

```python
dfc1['Sal'] = 16846,16160,11513,46464,79465
```

### Merge

```python
pd.merge(dfc1,dfc2,on='EMP_ID',how='left')
```
