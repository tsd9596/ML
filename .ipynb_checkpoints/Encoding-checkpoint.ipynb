- ML Model always deals with numerical values,If there;s categorical data we have convert that categorical data into Numerical

###### Types of Encoding

- get dummies
- One hot encoding
- Ordinal encoding (Independant Feature)
- Label encoding (Target Feature)
- Mean encoding

```python
import pandas as pd
import numpy as np
```

### Ordinal Encoding

```python
df = pd.read_csv(r"C:\Users\tsd95\CSV Files\Model\tips.csv")
```

```python
df
```

```python
df.columns
```

```python
e = df[[col for col in df.columns if col in ['sex', 'smoker', 'day', 'time']]]
```

```python
e
```

```python
e['sex'].unique()
```

```python
e['smoker'].unique()
```

```python
e['day'].unique()
```

```python
e['time'].unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ob = OrdinalEncoder(categories=[['Female', 'Male'],['No', 'Yes'],['Sun', 'Sat', 'Thur', 'Fri'],['Dinner', 'Lunch']])
```

```python
ob.fit(e)
```

```python
ob.transform(e)
```

```python
pd.DataFrame(ob.transform(e),columns=['Ensex','Ensmoker','Enday','Entime'])
```

```python
dfF = pd.concat([df,pd.DataFrame(ob.transform(e),columns=['Ensex','Ensmoker','Enday','Entime'])],axis=1)
```

```python
dfF
```

```python
dfF.drop(['sex', 'smoker', 'day', 'time'],axis=1,inplace=True)
```

```python
dfF
```

### One Hot Enoder

```python
df
```

```python
e
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ol = OneHotEncoder(drop='first')
```

```python
ol.fit(e)
```

```python
e1
```

```python
pd.DataFrame(e1,columns=['Female', 'Male','No', 'Yes','Dinner', 'Lunch'])
```

```python
Df = pd.concat([pd.DataFrame(e1,columns=['Female', 'Male','No', 'Yes','Dinner', 'Lunch']),df],axis=1)
```

```python
Df.drop(['sex', 'smoker', 'day', 'time'],axis=1,inplace=True)
```

### Label Encoder

```python
dfl = pd.read_csv(r"C:\Users\tsd95\CSV Files\Model\winequalityN.csv")
```

```python
dfl
```

```python
el = dfl['type']
```

```python
el.value_counts()
```

### Get Dummies

```python
dc = pd.read_csv(r"C:\Users\tsd95\CSV Files\Model\CarPrice_Assignment.csv")
```

```python
dc
```

```python
dc.columns
```

```python
dc.describe().T
```

```python
cc = dc[[col for col in dc.columns if col in ['fueltype', 'aspiration','doornumber', 'carbody', 'drivewheel', 'enginelocation', 'fuelsystem',]]]
```

```python
cc
```

```python
pd.get_dummies(cc)
```

```python
dc
```

### Mean Encoder

```python
dc['price'].groupby(dc['CarName']).mean()
```

```python
dc['CarName'] = dc['CarName'].map(dc['price'].groupby(dc['CarName']).mean())
```

```python
dc
```

- To Reduce Curs Of Dimentionality;We use Mean Encoding.
