- There are two ways to handling missing values.

- Deleting - dropna()
- Imputing - fillna()

### Imputation Of Null Values

- Important feature but missing.
- 20% Data Missing - Use Stastical Method.
- 80% Data Missing - Delete

#### Use Mean, Median and Mode

- For Numerical:   
   1.Mean : When there is no outliers.
   2.Median : Even if there is Outliers.
        
- For Categorical:
    1.Mode
