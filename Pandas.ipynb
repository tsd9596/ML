### loc and iloc

#### loc

- It is label based selecting method,which means we have to pass name of column which you want to select.
- This method includes the last element of range passed in it.
- Range excluded

```python
df.loc[10:20,'Gender':'Lunch']
```

#### Merge

```python
dfm1 = pd.DataFrame({'ID':[19,26,31,49,55],
                     'Name':['Vishnu','Komal','Nayan','Mayur','Ketan'],
                     'City':['Pune','Jaipur','Hydrabad','Delhi','Thane']})
```

```python
dfm1
```

```python
dfm2 = pd.DataFrame({'ID':[15,19,31,43,55],
                     'Product Name':['Cloths','Furniture','Electronic','Glosary','Shoes'],
                     'Price':[28486,14881,20506,50560,55500]})
```

```python
dfm2
```

```python
dfMer = pd.merge(dfm1,dfm2,on='ID')    #on=''........on which parameter you want merge
```

```python
dfMer               #common values merged
```

```python
#how is just like join in SQL
#By default .....how='inner'

dfMer = pd.merge(dfm1,dfm2,how='left')
```

```python
dfMer
```

```python
dfMer = pd.merge(dfm1,dfm2,on='ID',indicator=True,suffixes=('right','left'),how='left')
```

```python
dfMer
```
